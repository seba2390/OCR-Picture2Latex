```python
commands = []
with open('commands.txt', 'r') as File:
    for line in File:
        if line[:-1] not in commands and '[' not in line and '{' not in line and line != '\\':
            commands.append(line[:-1])
```

```python
with open('Unique_commands.txt', 'w') as File:
    for command in list(commands):
        File.write(command)
        File.write('\n')
```
